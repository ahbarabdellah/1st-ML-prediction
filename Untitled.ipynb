# My first machine learning prediction

```python
# load data
import pandas as pd
db= pd.read_csv('music.csv')
db
```

### Clean data and separate them in entry data X and predictions data Y

```python
#entry data X
X = db.drop(columns=['genre'])
X
```

```python
# predictions Y
Y = db['genre']
```

### use a decision tree algorithme from sklearn to train our model

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
# train model
model.fit(X,Y)
```

```python
# make a predicton a bout data doesen't exist in the sample what we train
age = int(input('Enter age :'))
sex = int(input('are you male enter 1/ if you femal entre 0'))
model.predict([[age,sex]])
```
